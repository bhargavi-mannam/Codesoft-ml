```python
data = data[['v1', 'v2']]
data.columns = ['label', 'message']

# Encode the labels (ham=0, spam=1)
data['label'] = data['label'].map({'ham': 0, 'spam': 1})

# Display the first few rows of the cleaned dataset
data.head()
```

# Splitting the Data

**In this step, we split the dataset into training and testing sets. This is a crucial part of preparing the data for machine learning, as it helps to evaluate the model's performance on unseen data. The training set is used to train the machine learning model, while the testing set is used to assess how well the model generalizes to new, unseen data.**

```python
from sklearn.model_selection import train_test_split

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(data['message'], data['label'], test_size=0.2, random_state=42)
```

**train_test_split:** *The train_test_split function is imported from the sklearn.model_selection module. This function is used to split the dataset into training and testing sets.*

**Specifying Features and Labels:** *We use the message column as the features (stored in X) and the label column as the labels (stored in y). In this case, X will be data['message'] and y will be data['label'].*

**Splitting the Data:**

*X_train and X_test contain the training and testing sets of the features (text messages), respectively.*

*y_train and y_test contain the training and testing sets of the labels, respectively.*

*test_size=0.2 indicates that 20% of the data will be used for testing, while the remaining 80% will be used for training.*

*random_state=42 ensures that the split is reproducible. Using the same random state ensures that the same split is obtained every time the code is run, which is useful for consistency and debugging.italicized text italicized text*

# Vectorizing the Text Data

**In this step, we are converting the text data into numerical format that can be used by machine learning algorithms. This is achieved using the CountVectorizer from the**

```python
from sklearn.feature_extraction.text import CountVectorizer

# Initialize the CountVectorizer
vectorizer = CountVectorizer()
# Fit and transform the training data, transform the testing data
X_train_vec = vectorizer.fit_transform(X_train)
X_test_vec = vectorizer.transform(X_test)
```

# Training a Multinomial Naive Bayes Classifier

**In this step, we are training a machine learning model using the Multinomial Naive Bayes algorithm. This classifier is particularly well-suited for discrete data, such as word counts obtained from text data.**

```python
from sklearn.naive_bayes import MultinomialNB

# Initialize and train the Multinomial Naive Bayes classifier
model = MultinomialNB()
model.fit(X_train_vec, y_train)
```

* *This line imports the MultinomialNB class from the sklearn.naive_bayes
module.*

* *Here, we create an instance of the MultinomialNB class. This instance will represent our Multinomial Naive Bayes classifier.*

* *The fit method is called on the classifier instance model with the training data X_train_vec and the corresponding labels y_train.*

* *The fit method trains the Multinomial Naive Bayes classifier on the training data. This involves calculating the probabilities of each word given each class (spam or ham) and the prior probabilities of each class. These probabilities are then used for making predictions on new data.*




# Evaluating the Model

**In this step, we evaluate the performance of the trained Multinomial Naive Bayes classifier using the test data. We will measure the accuracy of the model and generate a detailed classification report.**

```python
from sklearn.metrics import accuracy_score, classification_report

# Make predictions on the test data
y_pred = model.predict(X_test_vec)

# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
report = classification_report(y_test, y_pred)

# Print the results
print(f'Accuracy: {accuracy}')
print('Classification Report:')
print(report)
```

* *This line imports the accuracy_score and classification_report functions from the sklearn.metrics module. These functions will help us evaluate the performance of our classifier.*

* *The predict method of the trained model is called with the test data X_test_vec.*

* *This method uses the trained classifier to predict the labels (spam or ham) for each message in the test data.
The predicted labels are stored in the variable y_pred*
* *The accuracy_score function calculates the accuracy of the model by comparing the true labels (y_test) with the predicted labels (y_pred). Accuracy is the proportion of correctly classified messages.*

* *The classification_report function generates a detailed report showing the precision, recall, F1-score, and support for each class (spam and ham). Precision is the proportion of true positives among all predicted positives, recall is the proportion of true positives among all actual positives, and F1-score is the harmonic mean of precision and recall.*

* **The achieved accuracy of approximately 98.55% indicates a highly effective classifier, with detailed metrics showing strong performance in both identifying spam and non-spam messages.**

#  Confusion Matrix

**This code generates a confusion matrix to visualize the performance of a classification model. The confusion matrix is a table that allows visualization of the performance of an algorithm, typically a supervised learning one. It's often used in classification to understand the performance of the classification model.**

```python
from sklearn.metrics import confusion_matrix
import seaborn as sns

# Generate the confusion matrix
cm = confusion_matrix(y_test, y_pred)

# Plot the confusion matrix
plt.figure(figsize=(6, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=['Ham', 'Spam'], yticklabels=['Ham', 'Spam'])
plt.title('Confusion Matrix')
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.show()
```

# ROC Curve

**This code calculates the Receiver Operating Characteristic (ROC) curve and plots it to visualize the performance of the binary classification model. The ROC curve is a graphical plot that illustrates the diagnostic ability of a binary classifier system as its discrimination threshold is varied.**

```python
from sklearn.metrics import roc_curve, roc_auc_score

# Calculate the ROC curve
fpr, tpr, thresholds = roc_curve(y_test, model.predict_proba(X_test_vec)[:, 1])

# Plot the ROC curve
plt.figure(figsize=(6, 6))
plt.plot(fpr, tpr, label=f'AUC = {roc_auc_score(y_test, y_pred):.2f}')
plt.plot([0, 1], [0, 1], 'k--')
plt.title('ROC Curve')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.legend(loc='lower right')
plt.show()
```

* *We use the roc_curve function from scikit-learn to compute the ROC curve. This function takes the true binary labels (y_test) and the predicted probabilities of the positive class (model.predict_proba(X_test_vec)[:, 1]).*

* *We plot the ROC curve using plt.plot(). The x-axis represents the false positive rate (FPR), and the y-axis represents the true positive rate (TPR).*

* *The label parameter in plt.plot() adds a label to the plot, including the area under the curve (AUC) score.*

* *The diagonal dashed line ('k--') represents the ROC curve of a random classifier.*

#  Generating a Word Cloud for Ham Data

**This code generates a word cloud visualization for the "ham" data. A word cloud is a graphical representation of word frequency in a text corpus, where the size of each word indicates its frequency of occurrence. In this case, we are creating a word cloud specifically for the "ham" messages in our dataset.**

```python
pip install wordcloud
```

```python
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt

# Filter the dataframe to include only ham messages
ham_data_wordcloud = data[data['label'] == 0]

# Combine all the text from ham messages into a single string
ham_text = " ".join(ham_data_wordcloud['message'].values.tolist())

# Define the word cloud
wordcloud = WordCloud(
    width=800,
    height=800,
    stopwords=STOPWORDS,
    background_color='black',
    max_words=800,
    colormap="Set1"
).generate(ham_text)

# Display the word cloud
plt.figure(figsize=(10, 10))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()

```
